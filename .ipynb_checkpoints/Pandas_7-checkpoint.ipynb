```python
missing_values = ["N/a","na",]
```

```python
x = pd.read_csv("nan.csv",na_values = missing_values)
```

```python
x.isnull()
```

```python
x.isnull().sum()
```

```python
x.isnull().any()
```

```python
# They all are true it means all the column have NaN values
```

```python
sns.heatmap(x.isnull(),yticklabels=False, annot = True)
```

```python
# White colour show the NaN values per coloumn
```

# Step 2 :Remove all the values 

```python
x
```

```python
x.dropna(how = "all") # If row contain all NaN values then its remove but, we dont have that row so,not drop any row
```

```python
x.dropna() # If any row contain minimum one NaN value than it drop
```

```python
x.dropna(axis = 1) # If any column have NaN value it drop
```

```python
x.dropna(axis = 1,how = "all") # If any column contain all NaN values then it drop
```

# Step 3 : Fill NaN values

```python
x.fillna(0)
```

```python
x.fillna(method = "ffill") # Forward fill
```

```python
x.fillna(method = "bfill") # Backward fill
```

```python
x.fillna({
    "col.1":4444,
    "col.2":1111,
    "col.3":2222,
    "col.4":3333
},inplace = True)
```
